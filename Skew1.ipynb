```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from scipy import stats
```

### Generat Data

```python
data1=[34,18,50,50,26,18,34,26,67,52,52,67]
data2=pd.read_csv('data/data.csv')['age']
data3=np.random.normal(size=200, scale = 5,loc = 0) 
data4=stats.bernoulli.rvs(size=100,p=0.5,loc = 25)
data5=stats.binom.rvs(size=200,p=0.5,n=100,loc=0)
data6=stats.multinomial.rvs(size=200, p=[0.5,0.5],n=250 )
data7=stats.poisson.rvs(size=200, mu=10,loc=160 )

print("Data1 :\n"  ,data1,"\n",100*"*")
print("Data2 :\n"  ,data2,"\n",100*"*")
print("Data3 :\n"  ,data3,"\n",100*"*")
print("Data4 :\n"  ,data4,"\n",100*"*")
print("Data5 :\n"  ,data5,"\n",100*"*")
print("Data6 :\n"  ,data6,"\n",100*"*")
print("Data7 :\n"  ,data7,"\n",100*"*")
```

### Data Skew  

```python
print("Skew: ",stats.skew(data3))
print("Skew: ",stats.skew(data4))
print("Skew: ",stats.skew(data5))
print("Skew: ",stats.skew(data6))
print("Skew: ",stats.skew(data7))
```

```python
plt.hist(x=data3, bins='auto',alpha=0.9, rwidth=0.95, facecolor='magenta')
plt.hist(x=data4, bins='auto',alpha=0.9, rwidth=0.95, facecolor='green',edgecolor='black', linewidth=2)
plt.hist(x=data5, bins='auto',alpha=0.9, rwidth=0.95, facecolor='blue')
plt.hist(x=data6, bins='auto',alpha=0.9, rwidth=0.95, facecolor='red')
plt.hist(x=data7, bins='auto',alpha=0.9, rwidth=0.95, facecolor='yellow')
plt.grid(axis='y', alpha=0.75)
plt.ylim(ymax=90)
plt.xlim(xmax=200)
plt.show()
```
